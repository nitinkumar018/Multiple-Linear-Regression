# 5.Apply Lasso and Ridge methods on the model.

Ridge and Lasso are the tuning methods which come over and above the linear regression algorithm. 
Linear Regression algorithm has 2 methods to improve the performance of the model: Feature selection and Dedicated approach. 

In Feature selection, all the significant columns(variables) are retained and insignificant columns are removed and then train the model to predict the Y. This will improve the performance of the model.

Dedicated approach has Ridge and Lasso techniques. These techniques remove the underfitting and overfitting of the model and then train the model to predict the Y. This will also improve the performance of the model because overfitting and underfitting decrease the performance of the model.

Ridge reduces the β value of the insignificant column so that the impact of that column to predict the Y can be reduced. 

Lasso forces the β value of the insignificant column to be zero so that impact of that column to predict the Y can be eliminated.
